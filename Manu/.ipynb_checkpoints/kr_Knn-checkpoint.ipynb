```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import sklearn.preprocessing as preprocessing
from sklearn.metrics import mean_absolute_error
```

```python
test = pd.read_csv("../input/tp2datostrain/test.csv", index_col = 0)
train = pd.read_csv("../input/tp2datostrain/train.csv", index_col = 0)
display(test.head())

train.info()
```

### Proceso el DF 

```python
features = ['tipodepropiedad', 'provincia', 'habitaciones', 'antiguedad',
             'gimnasio', 'usosmultiples', 'piscina', 'escuelascercanas',
       'centroscomercialescercanos',]

trainFiltrado = train[features].copy()
```

```python
features_categoricos = ["tipodepropiedad","provincia"]
for col in features_categoricos:
    trainFiltrado[col].astype("category")
categorias_OHE = pd.get_dummies(trainFiltrado[features_categoricos], dummy_na=True)
categorias_OHE.columns = categorias_OHE.columns.str.strip()
```

```python
train_sin_cat = trainFiltrado.drop(features_categoricos, axis=1)
train_OHE = pd.concat([train_sin_cat, categorias_OHE], axis=1)
train_OHE.head()
```

```python
print(train_OHE.shape)
columnasTrain = train_OHE.columns
train_OHE.isnull().any()
```

### Creo train - validation

```python
colPredecir = "precio"
X = train_OHE.copy()
y = train[colPredecir]
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=1)
```

#### Imputo nulos solo en el x_train y no en el de validacion

```python
X_train.fillna(0,inplace = True)
X_val.fillna(0, inplace = True)
```

### Creo el modelo

```python
from sklearn.neighbors import KNeighborsClassifier
numeroVecinos = 5
classifier = KNeighborsClassifier(n_neighbors=numeroVecinos)
```

#### Entreno

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_val)
```

```python
mean_absolute_error(y_val, y_pred)
```

```python
def RMSLE(actual, pred):
    return (np.mean((np.log(actual + 1) - np.log(pred + 1)) ** 2)) **.5
RMSLE(y_val, y_pred)
```

### Genero la salida

```python
colPredecir = "precio"
X = train_OHE.copy()
y = train[colPredecir]
X.fillna(0,inplace=True)
knn.fit(X, y)
```

```python
features = ['tipodepropiedad', 'provincia', 'habitaciones', 'antiguedad',
             'gimnasio', 'usosmultiples', 'piscina', 'escuelascercanas',
       'centroscomercialescercanos',]

testFiltrado = test[features].copy()

features_categoricos = ["tipodepropiedad","provincia"]
for col in features_categoricos:
    testFiltrado[col].astype("category")
categorias_OHE = pd.get_dummies(testFiltrado[features_categoricos], dummy_na=True)
categorias_OHE.columns = categorias_OHE.columns.str.strip()

test_sin_cat = testFiltrado.drop(features_categoricos, axis=1)
test_OHE = pd.concat([test_sin_cat, categorias_OHE], axis=1)
test_OHE.head()
```

```python
columnasTest = test_OHE.columns
for col in columnasTest:
    if col not in columnasTrain:
        print(col, end = ", ")
columnasAAgregar = []
for col in columnasTrain:
    if col not in columnasTest:
        print(col, end= ", ")
        columnasAAgregar.append(col)
```

```python
for col in columnasAAgregar:
    test_OHE[col] = 0
test_OHE.fillna(0, inplace=True)
print(test_OHE.shape)
predicciones = knn.predict(test_OHE)
```

```python
predicciones
```

```python
res = pd.DataFrame(predicciones, index=test.index, columns=['precio'])
res.reset_index(inplace=True)
res.columns = ["id", "target"]
display(res.head())
res.to_csv("../input/resultados-knn.csv", header=False) # RMSLE=1.0249284784393988 ?


# import the modules we'll need
from IPython.display import HTML
import base64

# function that takes in a dataframe and creates a text link to  
# download it (will only work for files < 2MB or so)
def create_download_link(df, title = "Download CSV file", filename = "knn-results.csv"):  
    csv = df.to_csv(index=False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

# create a link to download the dataframe
create_download_link(res)
```
